```python
from databricks.sdk import WorkspaceClient
import time

w = WorkspaceClient()
TABLE_NAME = "main.financial.extracted_kpis"

print("Checking refresh status...\n")

while True:
    response = w.quality_monitors.list_refreshes(TABLE_NAME)
    
    # Extract refreshes from response object
    refreshes = response.refreshes if hasattr(response, "refreshes") else []
    
    if not refreshes:
        print("No refresh found yet — waiting...")
        time.sleep(15)
        continue
    
    latest = refreshes[0]
    state  = str(latest.state)
    
    if "SUCCESS" in state:
        print("Refresh COMPLETE")
        print("Go to: Catalog → main → financial → extracted_kpis → Quality tab")
        break
    
    elif "FAILED" in state:
        print("Refresh FAILED")
        print(f"Reason: {getattr(latest, 'message', 'no message')}")
        break
    
    elif "PENDING" in state or "RUNNING" in state:
        print(f"Still running... ({state.replace('RefreshState.', '')})")
        time.sleep(15)
    
    else:
        print(f"Unknown state: {state} — waiting...")
        time.sleep(15)
```
